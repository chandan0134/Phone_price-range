```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('mobile_price_range_data (2) (1).csv')
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1] 
y=df.iloc[:,-1] 
print(x.shape)
print(y.shape)
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.head()
```

LOGISTIC REGRESSION

```python
from sklearn.preprocessing import StandardScaler   
stx=StandardScaler()
x_train=stx.fit_transform(x_train)
x_test=stx.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression  
m= LogisticRegression()  
m.fit(x_train, y_train)  
```

```python
y_pred= m.predict(x_test)  
```

```python
m_pred= m.predict([[750,0,1.5,1,10,1,40,1.2,200,3,20,500,1000,3000,10,10,7,1,1,1]])
print(m_pred)
```

KNN ALGO:

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
m1=KNeighborsClassifier(n_neighbors=15)
m1.fit(x_train,y_train)
```

```python
m1_pred= m1.predict([[750,0,1.5,1,10,1,40,1.2,200,3,20,500,1000,3000,10,10,7,1,1,1]])
print(m1_pred)
```
